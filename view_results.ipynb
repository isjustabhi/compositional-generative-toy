```python
from PIL import Image
import matplotlib.pyplot as plt

def show_results(epoch):
    gray_path = f"samples/progress/input_gray_epoch{epoch}.png"
    color_path = f"samples/progress/output_color_epoch{epoch}.png"

    gray = Image.open(gray_path)
    color = Image.open(color_path)

    plt.figure(figsize=(10, 5))
    
    plt.subplot(1, 2, 1)
    plt.title(f"Grayscale Input (Epoch {epoch})")
    plt.imshow(gray)
    plt.axis('off')

    plt.subplot(1, 2, 2)
    plt.title("Colorized Output")
    plt.imshow(color)
    plt.axis('off')

    plt.tight_layout()
    plt.show()
```

```python
show_results(5)
show_results(10)
show_results(15)
show_results(20)
show_results(25)
show_results(30)
show_results(35)
show_results(40)
show_results(45)
show_results(50)

```
